# Análise de Dados de Filmes exibidos no Brasil entre 2009 a 2018
Com base nos dados da Ancine (https://oca.ancine.gov.br)

```python
import pandas as pd

%matplotlib inline
```

```python
# Criando DataFrame
#Pulado as 2 linhas do arquivo
df = pd.read_excel('ancine.xlsx', skiprows=2, skipfooter=16)

df.head()
```

```python
#Criar um novo DataFrame sem os Relançamentos
df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
df_novo.dtypes
```

```python
df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
df_novo[df_novo['Renda (R$) no ano de exibição'] == 
        df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
df_novo.nlargest(10,'Renda (R$) no ano de exibição')
```

```python
df_novo[df_novo['Renda (R$) no ano de exibição'] == 
        df_novo['Renda (R$) no ano de exibição'].min()]
```
